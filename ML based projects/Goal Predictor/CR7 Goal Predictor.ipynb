## Data PreProcessing 

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer = imputer.fit(X_train)

X_train = pd.DataFrame(imputer.transform(X_train), columns= X_train.columns)
X_test = pd.DataFrame(imputer.transform(X_test), columns= X_test.columns)
submission_x = pd.DataFrame(imputer.transform(submission_x), columns= submission_x.columns)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_array = scaler.transform(X_train)
X_train = pd.DataFrame(data = X_train_array, columns= X_train.columns, index= X_train.index)

X_test_array = scaler.transform(X_test)
X_test = pd.DataFrame(data = X_test_array, columns= X_test.columns, index= X_test.index)

submission_x_array = scaler.transform(submission_x)
submission_x = pd.DataFrame(data = submission_x_array, columns= submission_x.columns, index= submission_x.index)
```

## Training and Testing

```python
import xgboost as xgb
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn import tree
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import f1_score
```

```python
def train_classifier(clf, X_train, y_train):
    clf.fit(X_train, y_train)
    
def predict_labels(clf, features, target):
    y_pred = clf.predict(features)
    mae = mean_absolute_error(target, y_pred)
    
    
    #y_prob = clf.predict_proba(features) 
    
    return f1_score(target, y_pred), mae

def train_predict(clf, X_train, y_train, X_test, y_test):
    
    print("Training a {} using a training set size of {}".format(clf.__class__.__name__, len(X_train)))
    
    # Train the classifier
    train_classifier(clf, X_train, y_train)
    
    # Print the results of prediction for both training and testing
    f1, mae = predict_labels(clf, X_train, y_train)
    print("F1 score and mae for training set: {:.4f} , {:.4f}.".format(f1 , mae))
    
    f1, mae = predict_labels(clf, X_test, y_test)
    #print("y_pred_test : ")
    #print(y_pred_test)
    print("F1 score and mae for test set: {:.4f} , {:.4f}.".format(f1 , mae))
```

### Features Selection

```python
#features = ['range_of_shot','location_y','shot_basics','distance_of_shot']
```

```python
features = ['range_of_shot','location_x','shot_basics', 'type_of_shot_eff','remaining_min','location_y','distance_of_shot.1','remaining_sec']
```

## Model Selection

```python
clf_A = LogisticRegression()
clf_B = SVC()
clf_C = xgb.XGBClassifier()
clf_D = tree.DecisionTreeClassifier()
clf_E = RandomForestClassifier()
train_predict(clf_A, X_train[features], y_train, X_test[features], y_test)
print()
train_predict(clf_B, X_train[features], y_train, X_test[features], y_test)
print()
train_predict(clf_C, X_train[features], y_train, X_test[features], y_test)
print()
train_predict(clf_D, X_train[features], y_train, X_test[features], y_test)
print()
train_predict(clf_E, X_train[features], y_train, X_test[features], y_test)
print()
```

## Hyper Parameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer


# parameters to tune
parameters = { 'learning_rate' : [0.1],
               'n_estimators' : [1000, 1200],
               'max_depth': [4, 6, 10],
               'min_child_weight': [4, 6],
               'gamma':[0.0, 0.4],
               'subsample' : [0.8],
               'colsample_bytree' : [0.8],
               'objective' : ['binary:logistic'],
               'scale_pos_weight' : [1],
               'reg_alpha':[1e-5],
                'seed' : [42]
             }  

#  Initialize the classifier
clf = xgb.XGBClassifier(seed=2)

# Make an f1 scoring function using 'make_scorer' 
f1_scorer = make_scorer(f1_score,pos_label= 1)

# Perform grid search on the classifier using the f1_scorer as the scoring method
grid_obj = GridSearchCV(clf,
                        scoring=f1_scorer,
                        param_grid=parameters,
                        cv=5)

# Fit the grid search object to the training data and find the optimal parameters
grid_obj = grid_obj.fit(X_train[features],y_train)

# Get the estimator
clf = grid_obj.best_estimator_
print(clf)


# Report the final F1 score for training and testing after parameter tuning
f1, acc = predict_labels(clf, X_train[features], y_train)
print("F1 score and mae for training set: {:.4f} , {:.4f}.".format(f1 , acc))
    
f1, acc = predict_labels(clf, X_test[features], y_test)
print("F1 score and mae test set: {:.4f} , {:.4f}.".format(f1 , acc))
```
